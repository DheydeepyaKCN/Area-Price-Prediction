```python
import pandas as pd
```

```python
df=pd.read_csv("areaprice071.csv")
df
```

```python
x=df[["area"]]     #independent variable
x
```

```python
type(x)
```

```python
y=df.price      #dependent variable
y
```

```python
type(y)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.title("Area Vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y,color="red")
plt.show()
```

```python
#since only 1 independent and dependent value then use simple linear regression
```

```python
pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
#x needs be reshaped or in 2d or dataframe format
```

```python
model.fit(x,y)          #fit calculates m,c values needed for LR
```

```python
m1=model.coef_    #value of m
```

```python
c1=model.intercept_  #value of c
```

```python
y1=m1*5000+c1
```

```python
y1   #dont need to do manually you have function for it
```

```python
model.predict([[5000]])
```

```python
model.predict([[4000]])
```

```python
pred_y= model.predict(x)
```

```python
y
```

```python
df
```

```python
plt.title("Area Vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y,color="red")
plt.plot(x,pred_y,color="green")
plt.show()
```
